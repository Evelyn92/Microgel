```python
def calc_mue(M):
    """
     Args:
         M: point cloud
    returns mean of a point cloud
    """
    return np.mean(M, axis=0)

def calc_dist(p, mue): 
    """    
    At first, minus mue from point clouds, then calculate square 2 of them
     Args:
         p:   point cloud
         mue: mean of point cloud
    
    returns: calculated distance
    """
    return np.linalg.norm(p-mue, 2)

def norm_points(DATA, do_normalize):
    processed_points = []      
    points  = DATA
    confidence = 0.95

    if do_normalize:
        Inlaier_points = []
        mue            = calc_mue(points)# cal mean of a point cloud
    #     print(mue.shape)
    #     print('mue', mue)
        sorted_dists   = sorted([calc_dist(p, mue) for p in points])
        XY             = [(0, 0)]
        t              = len(sorted_dists)
        for j, d in enumerate(sorted_dists):
            XY.append(( d, 1.*j/t ))
            if 1.*j/t >= confidence:
                break
        sorted_idx     = sorted([(i, my_utils.calc_dist(p, mue)) for i, p in enumerate(points)], key= lambda d:(d[1], d[0]))
        inlaier        = [d[0] for d in sorted_idx[:j]]
        inlaier_points = [points[j] for j in inlaier]
        points         = np.array(inlaier_points)
    #     print(points.shape)
        points        -= mue
    
    return points
```

```python
def get_dataset(DATA_DIR,data_type, sample_num=4, tmp_num=6,  TEMP=['21C', '33C', '35C','38C','43C','53C'], load_NIPAM = True):
    selected_polymer = "NiPAM" if load_NIPAM == True else "NiPMAM"
    dataset_dict = {'Temperature':TEMP , 'selected_polymer':selected_polymer}
    for tmp_inx in range(1,tmp_num+1):
     # tmp_inx = 2 
        tmp_info = str(TEMP[tmp_inx-1])
        print(tmp_info)
        fig = plt.figure(figsize=(40,10))
        point_list = []
        for sample_inx in range(1,sample_num+1):
            # sample_inx = 1
            sample_file = 'sample_'+str(sample_inx)+data_type+str(tmp_inx)
            sample_pth = os.path.join(DATA_DIR, sample_file+'.3d')
            info_file = 'sample_'+str(sample_inx)+'info'+str(tmp_inx)+".3d"
            info_pth = os.path.join(DATA_DIR, info_file)

            sample_file = os.path.join(sample_pth)
            info_file = os.path.join(info_pth)

            DATA = pd.read_csv(sample_file, delimiter = "\t", header=None)
            INFO = pd.read_csv(info_file, delimiter = "\t", header=None)

            DATA = DATA.to_numpy()[:,:3]#only extract the location information
            INFO = INFO.to_numpy()
            DATA  = DATA  * (INFO[1,:]-INFO[2,:]) + INFO[2,:]+ INFO[0,:];
            DATA[:,2] = 1-DATA[:,2]


            DATA = DATA.astype(np.float)
            do_normalize = True
            DATA = norm_points(DATA, do_normalize)
            point_list.append(DATA)
            
        dataset_dict[tmp_info] = point_list
    return dataset_dict   

def depict_pdf(dataset_dict, sample_per_class = 4, TEMP=['21C', '33C', '35C','38C','43C','53C'], colors = ['g', 'b', 'k', 'y', 'm', 'c']):
    """
    This function receives a list of point clouds for a class(PEG,ODS,FOCTS,GLAS) and draws 
    probability density function (PDF). Also it depicts mean of each PDF in red.
    In each row draws 4 point clouds. 
    colors contain 6  diffrent colors. 
    
    
     Args:
         PC_list: a list of point clouds
         colors: list of 6 colors for depicting each row    
         class_label: class label
    """
    i = -1
    for tp in TEMP:
        fig = plt.figure(figsize=(10*sample_per_class,10))
        i = i+1
        pcs = dataset_dict[tp]
        sample_per_class = sample_per_class if (len(pcs)>sample_per_class) else len(pcs)
        for idx in range(sample_per_class):
            points = pcs[idx]
            ax1 = fig.add_subplot(1, sample_per_class, idx+1)            
            mue = calc_mue(points)
            sorted_dists = sorted([calc_dist(p, mue) for p in points])
            x_mean = np.mean(sorted_dists)
            XY = [(0, 0)]
            t = len(sorted_dists)
            for j, d in enumerate(sorted_dists):
                XY.append((d,1.*j/t))
            x, y = [xy[0] for xy in XY], [xy[1] for xy in XY]
            x = x/np.max(x)*100
            
            ax1.set_xlim((0,300))
            ax1.set_ylim((0,0.014))
            ax1.hist(sorted_dists, density=True, bins=20, range=[0,300])
            hist = np.histogram(sorted_dists, density=True, bins=20, range=[0,300])
            ax1.set_title('PDF, Temperature:{}, #:{}'.format(tp, idx))
            ax1.set_xlabel('Distance')
            ax1.set_ylabel('Probablity')
            ax1.legend()
            ax1.axvline(x_mean, color='red', linestyle='dashed', linewidth=1)
        plt.show()
    return hist
```

```python
package_name = 'seaborn'
spec = importlib.util.find_spec(package_name)
if spec is None:
    print(package_name +" is not installed")
    !pip install seaborn
else:
    print(package_name +" is installed")

import seaborn as sns
import matplotlib.colors as mcolors

color_css= mcolors.CSS4_COLORS
def depict_pdf_overlap(dataset_dict, sample_per_class = 4, TEMP=['21C', '33C', '35C','38C','43C','53C'], colors = [color_css['dimgray'], color_css['dimgrey'], color_css['gray'], color_css['grey'], color_css['darkgray'], color_css['darkgrey']]):
    """
    This function receives only the average samples so the sample_per_class should be 1
    It depicts the probability density function (PDF). Also it depicts mean of each PDF in red.
    It also return the values of the histogram for each temperature
    https://stackoverflow.com/questions/18449602/create-stacked-histogram-from-unequal-length-arrays
    """
    i = -1
    
    fig = plt.figure(figsize=(10*sample_per_class,10))
    ax1 = fig.add_subplot(1, 1, 1) 
    dist_list=[]
    prob_list=[]
    for tp in TEMP:
        i = i+1
        pcs = dataset_dict[tp]
#         sample_per_class = sample_per_class if (len(pcs)>sample_per_class) else len(pcs)
#         for idx in range(sample_per_class):
        idx = 0
        points = pcs[idx]
                   
        mue = calc_mue(points)
        sorted_dists = sorted([calc_dist(p, mue) for p in points])
        prob, bins = np.histogram(sorted_dists, density=True, bins=15, range=[0,300])
        ax1.hist(sorted_dists, density=True, bins=15, range=[0,300], color=colors[0], rwidth=0.9,alpha=0.5)
#         colors[5-idx]
        prob_list.append(prob)
        dist_list.append(sorted_dists)
#        fit the distribution: https://stackoverflow.com/questions/7805552/fitting-a-histogram-with-python
        x_mean = np.mean(sorted_dists)
        var = np.var(sorted_dists)
        pdf_x = np.linspace(np.min(sorted_dists),np.max(sorted_dists),100)
        pdf_y = 1.0/np.sqrt(2*np.pi*var)*np.exp(-0.5*(pdf_x-x_mean)**2/var)
#         XY = [(0, 0)]
#         t = len(sorted_dists)
#         for j, d in enumerate(sorted_dists):
#             XY.append((d,1.*j/t))
#         x, y = [xy[0] for xy in XY], [xy[1] for xy in XY]
#         x = x/np.max(x)*100
#         xx = bins[0:len(prob)]+10
        
        ax1.plot(pdf_x,pdf_y,'--')
    
    print(len(dist_list))
    ax1.set_xlim((0,300))
    ax1.set_ylim((0,0.014))

#     colors = sns.color_palette("tab20c")[14:20]
#     print(colors)
#     ax1.hist(dist_list, density=True, bins=15, range=[0,300], stacked=True, color=colors, rwidth=0.9)
    
    ax1.set_title('PDF, Temperature:{}, #:{}'.format(tp, idx))
    ax1.set_xlabel('Distance')
    ax1.set_ylabel('Probablity')
    ax1.legend(loc = 'upper right')
#     ax1.axvline(x_mean, color='red', linestyle='dashed', linewidth=1)

# https://matplotlib.org/stable/gallery/statistics/histogram_features.html

    plt.show()
    return
```
